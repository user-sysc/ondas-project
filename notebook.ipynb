# Simulación del Espectro Electromagnético
Este notebook simula el espectro electromagnético mostrando distintas bandas de radiación (Rayos Gamma, Rayos X, Ultravioleta, Luz Visible, Infrarrojo, Microondas y Ondas de Radio).


## El Espectro Electromagnético

Podemos clasificar y ordenar las ondas electromagnéticas de acuerdo a sus diferentes longitudes de onda y frecuencias; llamamos a esta clasificación "el espectro electromagnético". Abarca desde las ondas de radio, que tienen longitudes de onda muy largas y frecuencias bajas, hasta los rayos gamma, que tienen longitudes de onda extremadamente cortas y frecuencias muy altas.

### Clasificación del Espectro Electromagnético

El espectro se clasifica en diferentes regiones:

- **Ondas de Radio**: Son las ondas de menor frecuencia y mayor longitud de onda. Se utilizan en aplicaciones de comunicación como radio, televisión y redes inalámbricas(Wi-fi).
- **Microondas**: Son de menor longitud de onda que las ondas de radio y se utilizan en tecnologías como el radar, hornos microondas y comunicaciones satelitales.
- **Infrarrojo (IR)**: Estas ondas son emitidas por objetos en forma de calor y se emplean en aplicaciones como cámaras térmicas y controles remotos.
- **Luz Visible**: Es la única parte del espectro que el ojo humano puede detectar. Su rango de longitud de onda va de aproximadamente 400 a 700 nanómetros.
- **Ultravioleta (UV)**: Con longitudes de onda más cortas que la luz visible, las ondas UV son emitidas por el sol y pueden causar efectos biológicos, como el bronceado y quemaduras.
- **Rayos X**: Estas ondas son muy energéticas y penetrantes, lo que permite su uso en imágenes médicas y seguridad.
- **Rayos Gamma**: Son las ondas de mayor frecuencia y menor longitud de onda, producidas en procesos nucleares. Su alta energía permite aplicaciones en medicina (radioterapia) y en la observación de fenómenos cósmicos.

El espectro electromagnético se descubrió a raíz de los experimentos y los aportes del británico James Maxwell, quien descubrió la presencia de las ondas electromagnéticas y formalizó las ecuaciones de su estudio (conocidas como las ecuaciones de Maxwell).



Como podemos ver, el espectro visible —es decir, la luz que podemos ver con nuestros ojos— es tan solo una pequeña fracción de las diferentes clases de radiación que existen. A la derecha del espectro visible, encontramos las clases de energía que son menores en frecuencia (y por lo tanto mayores en longitud de onda) que la luz visible. Estas clases de energía incluyen los rayos infrarrojos (IR) (ondas de calor emitidas por los cuerpos térmicos), las microondas y las ondas de radio. Estos tipos de radiación nos rodean constantemente; no son dañinos, pues sus frecuencias son muy bajas. Las ondas de baja frecuencia tienen poca energía, y por lo tanto no son peligrosas para nuestra salud.

A la izquierda de espectro visible, encontramos los rayos ultravioleta (UV), los rayos X y los rayos gamma. Estas clases de radiación son dañinas para los organismos vivos, pues tienen frecuencias extremadamente altas (y por lo tanto, mucha energía). Es por esta razón que usamos loción bloqueadora en la playa (para bloquear los rayos UV provenientes del sol) y que, para prevenir que los rayos X penetren otras áreas del cuerpo distintas de la que requiere visualizarse, un técnico de rayos X coloca una placa de plomo sobre nosotros. Los rayos gamma son los más dañinos, pues son los más altos en frecuencia y en energía. Afortunadamente, nuestra atmósfera absorbe los rayos gamma que provienen del espacio, y así nos protege del daño.


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Configuración de las bandas del espectro electromagnético
spectro = {
    "Rayos Gamma": (1e-12, 1e-10),
    "Rayos X": (1e-10, 1e-8),
    "Ultravioleta": (1e-8, 4e-7),
    "Luz Visible": (4e-7, 7e-7),
    "Infrarrojo": (7e-7, 1e-3),
    "Microondas": (1e-3, 1e-1),
    "Ondas de Radio": (1e-1, 1e3)
}

# Crear una figura para la gráfica
fig, ax = plt.subplots()
plt.subplots_adjust(left=0.1, bottom=0.25)
ax.set_xscale("log")
ax.set_xlim(1e-12, 1e3)
ax.set_ylim(0, 1)
ax.set_xlabel("Longitud de onda (m)")
ax.set_ylabel("Intensidad relativa")

# Colores para las diferentes bandas
colors = ["purple", "blue", "cyan", "green", "yellow", "orange", "red"]
for i, (tipo, (min_lambda, max_lambda)) in enumerate(spectro.items()):
    ax.fill_betweenx([0, 1], min_lambda, max_lambda, color=colors[i], alpha=0.5, label=tipo)

# Mostrar leyenda y gráfica
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Velocidad de la luz en m/s
c = 3e8

# Configuración de las bandas del espectro electromagnético
spectro = {
    "Rayos Gamma": (1e-12, 1e-10, "purple"),
    "Rayos X": (1e-10, 1e-8, "blue"),
    "Ultravioleta": (1e-8, 4e-7, "cyan"),
    "Luz Visible": (4e-7, 7e-7, "green"),
    "Infrarrojo": (7e-7, 1e-3, "yellow"),
    "Microondas": (1e-3, 1e-1, "orange"),
    "Ondas de Radio": (1e-1, 1e3, "red")
}

# Crear una figura con subgráficas para cada tipo de onda
fig, axes = plt.subplots(len(spectro), 1, figsize=(8, 12))
fig.suptitle("Espectro Electromagnético: Longitud de Onda y Frecuencia por Tipo de Onda", fontsize=16)

for ax, (tipo, (min_lambda, max_lambda, color)) in zip(axes, spectro.items()):
    # Rango de longitudes de onda para esta banda
    lambdas = np.logspace(np.log10(min_lambda), np.log10(max_lambda), 100)
    # Calcular la frecuencia correspondiente a cada longitud de onda
    frecuencias = c / lambdas
    
    # Graficar longitud de onda vs frecuencia para esta banda
    ax.plot(lambdas, frecuencias, color=color)
    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.set_xlim(min_lambda, max_lambda)
    ax.set_xlabel("Longitud de onda (m)")
    ax.set_ylabel("Frecuencia (Hz)")
    ax.set_title(f"{tipo}")
    ax.grid(True, which="both", linestyle="--", linewidth=0.5)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

La razón por la cual todas las gráficas (rayos gamma, rayos X y ultravioleta) se ven con una forma similar en términos de pendiente y dirección es porque todas estas ondas electromagnéticas siguen la misma **relación inversa** entre **frecuencia** y **longitud de onda**. Esta relación está dada por la ecuación:

$$
f = \frac{c}{\lambda}
$$

donde:

- ${f}$ es la **frecuencia**,
- ${\lambda}$ es la **longitud de onda**,
- ${c}$ es la **velocidad de la luz** ${(3 \times 10^8) m/s}$, que es constante para todas las ondas electromagnéticas.

**Relación Inversa**: Como la frecuencia y la longitud de onda están inversamente relacionadas, cuando la longitud de onda aumenta, la frecuencia disminuye proporcionalmente, y viceversa. Esto significa que en la gráfica, a medida que nos movemos hacia la derecha en el eje de longitud de onda (valores más grandes), los valores de frecuencia en el eje vertical disminuyen.

**Escala Logarítmica**: En estas gráficas, tanto el eje de la longitud de onda como el de la frecuencia están en escala logarítmica. Esta escala permite representar de manera proporcional el amplio rango de valores de longitud de onda y frecuencia en el espectro electromagnético.

## Cuantización de la energía y la naturaleza dual de la luz
A finales del siglo diecisiete, el físico holandés Christiaan Huygens fue el primero en describir la naturaleza ondulatoria de la luz, Sin embargo, esto cambió en 1900, cuando el físico Max Planck comenzó a estudiar cuerpos negros —cuerpos que se calientan hasta que empiezan a brillar—.
Planck encontró que la radiación electromagnética emitida por un cuerpo negro no podía ser explicada por la física clásica, que postula que la materia puede absorber y emitir cualquier cantidad de radiación electromagnética.
La realidad, que descubrió Planck, es que la energía no es continua, sino que está cuantizada —es decir, que solo puede transferirse en "paquetes" individuales (o partículas) de tamaño hv—. Cada uno de estos paquetes de energía es conocido como cuanto (cuantos, en plural).
    
$$
h = 6.626 \times 10^{-34} \, \text{J} \cdot \text{s}
$$

$$
E = h \cdot f
$$

- \(E\): Energía del fotón (en julios, \(J\)).
- \(h\) = 6.626x10^-34 J*s: Constante de Planck.
- \(f\): Frecuencia de la onda electromagnética (en hertzios, \(Hz\)).
